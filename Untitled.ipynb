```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm

sns.set(style="whitegrid")

np.random.seed(123)
n = 150

interest_rates = np.random.uniform(0.5, 5.0, n)
share_prices = np.random.uniform(80, 200, n)

df = pd.DataFrame({
    'InterestRate': interest_rates,
    'SharePrice': share_prices
})

df_clean = df.dropna()
print(f"Dataset length after cleaning: {len(df_clean)}")

mean_interest = np.mean(df_clean['InterestRate'])
median_interest = np.median(df_clean['InterestRate'])
std_interest = np.std(df_clean['InterestRate'])
var_share_price = np.var(df_clean['SharePrice'])
range_share_price = np.ptp(df_clean['SharePrice'])

print("\nNumPy Operations:")
print(f"Mean Interest Rate: {mean_interest:.3f}")
print(f"Median Interest Rate: {median_interest:.3f}")
print(f"Std Dev Interest Rate: {std_interest:.3f}")
print(f"Variance Share Price: {var_share_price:.3f}")
print(f"Range Share Price: {range_share_price:.3f}")

corr_coef, p_value = stats.pearsonr(df_clean['InterestRate'], df_clean['SharePrice'])
print(f"\nPearson Correlation Coefficient: {corr_coef:.3f}, p-value: {p_value:.4f}")

X = sm.add_constant(df_clean['InterestRate'])
y = df_clean['SharePrice']
model = sm.OLS(y, X).fit()
print("\nStatsmodels Regression Summary:")
print(model.summary())

plt.figure(figsize=(8, 5))
sns.histplot(df_clean['InterestRate'], bins=15, kde=True, color='skyblue')
plt.title('Histogram of Interest Rates')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(6, 5))
sns.boxplot(x=df_clean['SharePrice'], color='lightgreen')
plt.title('Boxplot of Share Prices')
plt.xlabel('Share Price ($)')
plt.show()

sns.lmplot(x='InterestRate', y='SharePrice', data=df_clean, height=6, aspect=1.5, line_kws={'color': 'red'})
plt.title('Correlation Between Interest Rate and Share Price')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Share Price ($)')
plt.show()
```
